# Задача 1: сравнение предложений

## 1.1. Считываем данные из файла

```python
with open('sentences.txt', 'r') as file_sentences:
    data_sentences = list(file_sentences)
    str_sentences = str(data_sentences)
    print(data_sentences)
```

## 1.2. Разбиваем на строки

```python
for line in data_sentences:
    print(line)
```

## 1.3. Приводим строки к нижнему регистру

```python
lowercase_sentences = []
for i in range(len(data_sentences)):
    lowercase_sentences.append(data_sentences[i].lower())
```

```python
for line in lowercase_sentences:
    print(line)
```

## 1.4. Разбиваем текст на слова (токенизация)

```python
import re

words_sentences = []
for line in lowercase_sentences:
    words_sentences.append(re.split('[^a-z]', line))

for line in words_sentences:
    print(line)
```

## 1.5. Чистим от пробелов

```python
clean_sentences = [[] for i in range(len(words_sentences))]

i = 0
for line in words_sentences:
    for words in line:
        if words != '':
            clean_sentences[i].append(words)
    i = i + 1
```

```python
for line in clean_sentences:
    print(line)
```

## 1.6. Составляем список всех слов

```python
dict_sentences = {}

i = 0
for line in clean_sentences:
    for words in line:
        if words not in dict_sentences:
            dict_sentences[words] = i 
            i += 1
            
for item in dict_sentences:
    print (item, ': ', dict_sentences[item])
```

```python
#d - число различных слов в предложениях

d = len(clean_sentences)
print(d)
```

```python
#Посмотрим сколько у нас всего слов
n = len(dict_sentences)
print (n)
```

```python
#Посмотрим что это за слова
print (dict_sentences.keys())
```

## 1.7. Создаем матрицу

```python
import numpy as np

n = len(dict_sentences)
matrix = np.zeros((n, d))
        
print (matrix)
```

```python
#Посмотрим какого размера матрица должна быть
import pandas as pd
sentences_frame = pd.DataFrame(dict_sentences, range(len(clean_sentences)))
rows_count, columns_count = sentences_frame.shape
print (rows_count, columns_count)
```

```python
#Создаем пустую матрицу размером n * d
#Где n - число предложений, а d - число различных слов в предложениях

for i in range(rows_count):
    for j in range(columns_count):
        sentences_frame.iloc[i,j] = 0
        
sentences_frame
```

```python
#Заполняем матрицу где элемент имеет индекс (i,j)
#Дает количество вхождений j-го слова в i-е предложение
for i in range(len(clean_sentences)):
    for j in clean_sentences[i]:
        sentences_frame.loc[i, j] += 1
        
sentences_frame
```

## 1.7. Находим косинусное растояние  от предложения в самой первой строке до всех остальных

```python
from scipy.spatial.distance import cosine
```

```python
distance_from_first_sentence = []

for i in range(rows_count):
    distance_from_first_sentence.append(cosine(sentences_frame.loc[0], sentences_frame.iloc[i]))
    
print (distance_from_first_sentence)
```
